## Anotar Imagens
- Desenhar Linhas
- Desenhar Circulos
- Desenhar Retangulos
- Adicionar Texto

```python
# Bibliotecas necess√°rias
import cv2 as cv
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from IPython.display import Image

%matplotlib inline
```

```python
# Ler Imagem
image = cv.imread('./img/Apollo_11_Launch.jpg', cv.IMREAD_COLOR)

# Mostar Imagem
plt.imshow(image[:,:,::-1])
```

### Desenhar Linhas

```python
image_line = image.copy()

# imagem, inicio, fim, cor, grossura, tipo de linha
cv.line(image_line, (200, 100), (400, 100), (0, 255, 0), 5, lineType=cv.LINE_AA)

# Mostrar imagme
plt.imshow(image_line[:,:,::-1])
```


### Desenhar Circulos

```python
image_circle = image.copy()

# Image, centro, raio, cor, grossura, tipo de lina
cv.circle(image_circle, (900, 500), 100, (255, 0, 255), thickness=15, lineType=cv.LINE_AA)

# Mostrar imagem
plt.imshow(image_circle)
```

### Desenhar Retangulo

```python
image_rect = image.copy()

# image, inicio, fim, cor, grossura, tipo de linha
cv.rectangle(image_rect, (200, 100), (500, 500), (0, 255, 0), thickness=5, lineType=cv.LINE_8)

# Mostar imagem
plt.imshow(image_rect)
```

### Adicionar Texto

```python
image_text = image.copy()

text = "Apollo II, Saturn V Launch"
font_scale = 2.3
font_face = cv.FONT_HERSHEY_COMPLEX
font_color = (255, 0, 255)
thickness = 5

# Image, texto, inicio, fonte, tamanho, cor, grossura, tipo de linha
cv.putText(image_text, text, (150, 700), font_face, font_scale, font_color, thickness, cv.LINE_AA)


plt.imshow(image_text)
```
